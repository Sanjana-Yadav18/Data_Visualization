```python
#  Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
# dimensional space.
#                                      np.random.seed(30)
#                                      data = {
#                                          'X': np.random.uniform(-10, 10, 300),
#                                          'Y': np.random.uniform(-10, 10, 300),
#                                          'Z': np.random.uniform(-10, 10, 300)
#                                      }
#                                      df = pd.DataFrame(data)



import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=df['X'],
    y=df['Y'],
    z=df['Z'],
    mode='markers',
    marker=dict(
        size=5,
        color='skyblue',                # set color to an array/list of desired values
        opacity=0.8
    )
)])

# Set layout
fig.update_layout(
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
    ),
    title='3D Scatter Plot of Data Points'
)

# Show plot
fig.show()
```

```python
#  Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
#                              np.random.seed(20)
#                             data = {
#                            'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#                            'Day': np.random.choice(range(1, 31), 100),
#                            'Sales': np.random.randint(1000, 5000, 100)
#                             }
#                            df = pd.DataFrame(data) 


import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate Sales dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table for heatmap
sales_heatmap = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

# Create heatmap using Plotly
fig = go.Figure(data=go.Heatmap(
                   z=sales_heatmap.values,
                   x=sales_heatmap.columns,
                   y=sales_heatmap.index,
                   colorscale='Viridis'))

# Update layout
fig.update_layout(
    title='Variation in Sales Across Different Months and Days',
    xaxis=dict(title='Day'),
    yaxis=dict(title='Month')
)

# Show plot
fig.show()
```

```python
#  Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(underroot(x^2 + y^2)).
#                                    x = np.linspace(-5, 5, 100)
#                                   y = np.linspace(-5, 5, 100)
#                                   x, y = np.meshgrid(x, y)
#                                   z = np.sin(np.sqrt(x**2 + y**2))
#                                   data = {
#                                   'X': x.flatten(),
#                                   'Y': y.flatten(),
#                                   'Z': z.flatten()
#                                   }
#                                   df = pd.DataFrame(data)


import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z data
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout
fig.update_layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show plot
fig.show()
```

```python
# Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and 
# bubble size proportional to the population.
#                                      np.random.seed(25)
#                                      data = {
#                                           'Country': ['USA', 'Canada', 'UK',
#                                      'Germany', 'France'],
#                                           'Population':
#                                       np.random.randint(100, 1000, 5),
#                                           'GDP': np.random.randint(500, 2000,
#                                       5)
#                                       }
#                                       df = pd.DataFrame(data) using ploty

import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create bubble chart using Plotly
fig = go.Figure()

# Add trace for each country
for i, row in df.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['GDP']],
        y=[row['Population']],
        mode='markers',
        marker=dict(
            size=row['Population'],  # Bubble size proportional to population
            sizemode='area',
            sizeref=2.*max(df['Population'])/(50.**2),
            sizemin=4
        ),
        name=row['Country']
    ))

# Update layout
fig.update_layout(
    title='Bubble Chart: Population vs GDP',
    xaxis_title='GDP',
    yaxis_title='Population'
)

# Show plot
fig.show()
```
